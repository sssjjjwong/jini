```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('modu_csv/pima-indians-diabetes.csv',header=None)
df
```

```python
plt.figure(figsize=(12,12))

sns.heatmap(df.corr(), linewidths = 0.1,vmax = 0.5,cmap = plt.cm.gist_heat,linecolor = 'black',annot = True)
plt.show()
```
